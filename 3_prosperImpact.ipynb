# PROSPERIMPACT

This notebook calculates the dynamic impact as additional radiative forcing and saves the results in excel files.

environment: premise2 (theoretically every other)

Calculations are possible for methane, carbon dioxide and dinitrogen oxide. The calculations follow the IPCC assessment report 5, chapter 8. The calculations for dinitrogen oxide are slihtly simplified.

# Impulse response functions

```python
import math
```

```python
def prospRF(measd):
    t = 0
    reCO2 = 1.37e-05 #[W/(m²ppb)] entsp. AR5 Table 8A1
    reCH4 = 3.63e-4 #[W/(m²ppb)] entsp. AR5 Table 8A1
    reN2O = 3.00e-3 #[W/(m²ppb)] entsp. AR5 Table 8A1
    mbCO2 = 1.28301e-10 #[ppb/kg] 
    mbCH4 = 3.52827e-10 #[ppb/kg] 
    mbN2O = 1.22722e-10 #[ppb/kg] 
    f1f2 = 1 + 0.5 + 0.15 #entsp. AR5 8SM
    trfCO2 = []
    for t in range(measd):
        rfCO2 = mbCO2 * reCO2 * (0.2173 + 0.224 * math.exp(-(t/394.4)) + 0.2824 * math.exp(-(t/36.54)) + 0.2763 * math.exp(-(t/4.304)))
        t = t + 1
        trfCO2.append(rfCO2)
    trfCH4 = []
    for t in range(measd):
        rfCH4 = mbCH4 * reCH4 * f1f2 * math.exp(-(t/12.4))
        t = t + 1
        trfCH4.append(rfCH4)
    trfN2O = []
    for t in range(measd):
        rfN2O = mbN2O * reN2O * math.exp(-(t/121.0))
        t = t + 1
        trfN2O.append(rfN2O)
    radf={'CO2': trfCO2, 'CH4': trfCH4, 'N2O': trfN2O}
    return(radf)
```

```python
# time span for calculation of the radiative forcing
measd = 112
```

```python
#ensures, that values exist for the whole time span

def add_missing_years(df, years_range):
    missing_years = [year for year in years_range if year not in df['year'].values]
    missing_data = pd.DataFrame({'year': missing_years, 'process': df['process'][ultimo], 'CO2': df['CO2'][ultimo], 'CH4': df['CH4'][ultimo], 'N2O': df['N2O'][ultimo]})
    df = pd.concat([df, missing_data], ignore_index=True)
    return df
```

```python
pro = []
for em in df['process']:
    if em not in pro:
        pro.append(em)
        
print(pro)
```

```python
def convolvo(pro, df, radf, t0):
    radiat = {}
    for p in pro:
        e = 0
        print(p)
        for em in df:
            print('em', em)
            t = 0
            for ga in radf:
                if ga==em:
                    print('ga', ga)
                    cumrf = [0]
                    tempo = [t0]
                    z = 0
                    for m in df['process']:
                        if m == p:
                            zw = 0
                            for tau in range(t+1):
                                try:
                                    zw = zw + df[em][z-t+tau] * radf[ga][t-tau]
                                except:
                                    raise IndexError('calculated range of impulse response function too short! Change measd.')
                            #    print(z, t, tau)
                            cumrf.append(zw)
                            tempo.append(df['year'][z])
                            t = t+1
                        z = z+1
                    print(cumrf)
                    print(tempo)
                    radiat[p,ga]=cumrf
            e = e+1
    return(tempo, radiat)
```
